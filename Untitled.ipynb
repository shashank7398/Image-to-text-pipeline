```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
blurred = cv2.GaussianBlur(gray, (5, 5), 0)# GaussianBlur image 
```

```python
#get threshold image
thresh = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, 
                               cv2.THRESH_BINARY, 11, 2)
```

```python
processed_image_path = "processed_menu.jpg"
cv2.imwrite(processed_image_path, thresh)
print(f"Preprocessed image saved as: {processed_image_path}")
```

```python
# Extract Text using OCR
def extract_text(image_path):
    """Extracts text from an image using Tesseract OCR"""
    image = Image.open(image_path)
    extracted_text = pytesseract.image_to_string(image)
    return extracted_text.strip()
```

```python
def translate_text(text):
    """Detects language and translates text to English if needed"""
    detected_lang = detect(text)
    if detected_lang != "en":
        translator = Translator()
        translated_text = translator.translate(text, src=detected_lang, dest="en").text
        return translated_text
    return text
```

```python
# Structure Menu Using GPT-4
def structure_menu_with_gpt(menu_text):
    """Formats extracted menu text into structured JSON using OpenAI GPT"""
    prompt = f"""
    Given the following restaurant menu text, structure it into a JSON format with 'categories', 'items', and 'prices'.

    Menu:
    {menu_text}

    JSON Format:
    {{
      "categories": [
        {{
          "category_name": "Appetizers",
          "items": [
            {{"name": "Spring Rolls", "price": "$5.99"}},
            {{"name": "Garlic Bread", "price": "$4.50"}}
          ]
        }},
        {{
          "category_name": "Main Course",
          "items": [
            {{"name": "Grilled Chicken", "price": "$12.99"}},
            {{"name": "Pasta Alfredo", "price": "$10.50"}}
          ]
        }}
      ]
    }}
    """

    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "system", "content": "You are an AI that formats menus."},
                  {"role": "user", "content": prompt}]
    )

    try:
        structured_menu = json.loads(response["choices"][0]["message"]["content"])
        return structured_menu
    except json.JSONDecodeError:
        return {"error": "Failed to structure menu"}
```

```python
print("\nExtracting text from image...")
extracted_text = extract_text(processed_image_path)
print("\nExtracted Text:\n", extracted_text)
```

```python
structured_menu_json = structure_menu_with_gpt(extracted_text)
print("Structured JSON Output:\n", structured_menu_json)
```
